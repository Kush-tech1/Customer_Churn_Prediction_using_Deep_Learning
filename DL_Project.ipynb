```python
import pandas as pd
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data.head()
```

```python
data=data.drop(columns=['RowNumber','CustomerId','Surname','Geography','Gender'])
```

```python
data.head()
```

```python
input_data= data.iloc[:,:-1]
output_data= data.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
input_data = pd.DataFrame(ss.fit_transform(input_data), columns=input_data.columns )
```

```python
input_data
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2, random_state=42)
```

```python
X_train.shape
```

```python
import tensorflow
```

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
ann = Sequential()
```

```python
ann.add(Dense(6, input_dim=8, activation='relu'))
ann.add(Dense(4, activation='relu'))
ann.add(Dense(2, activation='relu'))
ann.add(Dense(1, activation='sigmoid'))
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ann.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
prd=ann.predict(X_test)
prd_data = []
for i in prd:
    if i[0] > 0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
prd=ann.predict(X_train)
prd_data1 = []
for i in prd:
    if i[0] > 0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
from sklearn.metrics import accuracy_score
print('test accurcy:',accuracy_score(y_test, prd_data)*100)
print("train accuracy",accuracy_score(y_train, prd_data1)*100)
```
